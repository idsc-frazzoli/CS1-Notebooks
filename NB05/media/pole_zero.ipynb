```python
import matplotlib.pyplot as plt

```

```python
def plot_system_pz(real_zero, pole1_real, pole1_img,pole2_real, pole2_img):
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 2)
    plt.scatter(real_zero, 0, color='blue', marker='o', facecolors='none', label='Zero')  # Display zero as negative

    plt.scatter(pole1_real, pole1_img, color='red', marker='x', label='Pole 1')
    plt.scatter(pole2_real, pole2_img, color='green', marker='x')
    #plt.scatter(0, 3, color='green', label='input', s=30, marker='s')

    plt.axhline(0, color='black', lw=0.5, ls='--')
    plt.axvline(0, color='black', lw=0.5, ls='--')
    plt.title('Pole-Zero Map')
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.grid()
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
plot_system_pz(1,3,-2,3,2)
```
